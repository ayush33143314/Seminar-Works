```python
name=r"C:\\Users\\ayush\\Desktop\\Final\\base\\data\\maze\\grid10.txt"
with open(name) as f:
    content = f.readlines()
name = [content[i][:-1].split() for i in range(len(content))]
numStates=0
numActions=4
start=''
end=''
discount=1
for i in range(len(name)):
    for j in range(len(name[0])):
        if name[i][j]=='0'  or name[i][j]=='2'or name[i][j]=='1' or name[i][j]=='3':
            numStates=numStates+1
        if name[i][j]=='2':
            start=str(i*len(name[0])+j)
        if name[i][j]=='3':
            end=str(i*len(name[0])+j)

file=open("C:/Users/ayush/desktop/Final/base/data/mdp/mdp2.txt","w")
file. truncate(0)
L='numStates '+str(numStates)
file.writelines(L)
L='\nnumActions '+str(numActions)
file.writelines(L)    
L='\nstart '+end
file.writelines(L)    
L='\nend '+start
file.writelines(L)  
index=0
for i in range(0,len(name)):
    for j in range(0,len(name[0])):
        if name[i][j]=='0' or name[i][j]=='2':
            count=0
            if name[i-1][j]=='0':
                count+=1
            if name[i+1][j]=='0':
                count+=1
            if name[i][j-1]=='0':
                count+=1
            if name[i][j+1]=='0':
                count+=1
            probability=0
            if count==1:
                probability=1
            if count==2:
                probability=0.5
            if count==3:
                probability=0.33
            if count==4:
                probability=0.25
            if name[i][j-1]=='0':
                L='\ntransition ',str(index),' 3 ',str(index-1),' 1 '+str(probability)
                file.writelines(L)    
            if name[i][j+1]=='0':
                L='\ntransition ',str(index),' 1 ',str(index+1),' 1 '+str(probability)
                file.writelines(L)
            if name[i-1][j]=='0':
                L='\ntransition ',str(index),' 0 ',str(index-len(name[0])),' 1 '+str(probability)
                file.writelines(L)
            if name[i+1][j]=='0':
                L='\ntransition ',str(index),' 2 ',str(index+len(name[0])),' 1 '+str(probability)
                file.writelines(L)
            if name[i][j-1]=='3':
                L='\ntransition ',str(index),' 3 ',str(index-1),' 100000000 '+str(probability)
                file.writelines(L)    
            if name[i][j+1]=='3':
                L='\ntransition ',str(index),' 1 ',str(index+1),' 100000000 '+str(probability)
                file.writelines(L)
            if name[i-1][j]=='3':
                L='\ntransition ',str(index),' 0 ',str(index-len(name[0])),' 100000000 '+str(probability)
                file.writelines(L)
            if name[i+1][j]=='3':
                L='\ntransition ',str(index),' 2 ',str(index+len(name[0])),' 100000000 '+str(probability)
                file.writelines(L)
            
            if name[i][j-1]=='2':
                L='\ntransition ',str(index),' 3 ',str(index-1),' -100000000 '+str(probability)
                file.writelines(L)    
            if name[i][j+1]=='2':
                L='\ntransition ',str(index),' 1 ',str(index+1),' -100000000 '+str(probability)
                file.writelines(L)
            if name[i-1][j]=='2':
                L='\ntransition ',str(index),' 0 ',str(index-len(name[0])),' -100000000 '+str(probability)
                file.writelines(L)
            if name[i+1][j]=='2':
                L='\ntransition ',str(index),' 2 ',str(index+len(name[0])),' -100000000 '+str(probability)
                file.writelines(L)
        index=index+1
L='\nmdptype episodic'
file.writelines(L)
L='\ndiscount 1.0'
file.writelines(L)    
file.close()
```
